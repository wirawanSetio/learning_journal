```python
hasil = []
i = 0
length = len(nba)

while(i<length):
    selisih = nba.iloc[i,10] - nba.iloc[i,14]
    hasil.append(selisih)
    i+=1

nba['diff_manual'] = hasil
```

```python

    
def match_result(diff):
    if diff > 0:
        return "menang"
    elif diff < 0:
        return "kalah"
    else:
        return 'imbang'
```

```python

keterangan = []

for diff in nba['diff_manual']:
    result = match_result(diff)
    keterangan.append(result)

nba['hasil_game'] = keterangan
```
